We can see that there are some really messy attributes, let's see if we can make sense of them

```python
english_2016.loc[0]['possession']
```

Looks like XML. Let's process it.

```python
import xml.etree.ElementTree as ET
tree = ET.fromstring(english_2016.loc[0]['possession'])
```

```python
for child in tree:
    print(child.tag,child.text)
    for second in child:
        print("\t",second.tag,second.text)
        for third in second:
            print("\t","\t",third.tag,third.text)
```

The "elapsed" seems to show how many minutes have elapsed from the start of the match and then homepos is a percentage of how much the home team has had the ball so far and awaypos the same for the away team. Possession can certainly be an important factor when training a model but not at all points of the match. It seems reasonable to extract the last possession number and use it during the training.

```python
#find the last occurence of "homepos" and look a few characters before that for the number
string = english_2016.loc[0]['possession']
index = string.rfind('homepos')
homepos = string[index-4:index-2]
int(homepos)
```

```python
newpos = []
for val in english_2016['possession']:
    index = val.rfind('homepos')
    homepos = val[index-4:index-2]
    newpos.append(int(homepos))
```

We only need the home team possession because the away team possession can easily be deduced from that and would provide no extra info for the neural net

```python
english_2016['possession'] = newpos
```
