```python
import pandas as pd
```

```python
data = pd.read_csv('data/english_2016.csv')
```

```python
data
```

```python
for col in data.columns:
    print(col)
```

```python
newpos = []
for val in data['possession']:
    index = val.rfind('homepos')
    homepos = val[index-4:index-2]
    newpos.append(int(homepos))
data['possession'] = newpos
```

```python
targets = []
for match in data.iterrows():
    if match[1]['home_team_goal']>match[1]['away_team_goal']:
        targets.append([1,0,0])
    if match[1]['home_team_goal']==match[1]['away_team_goal']:
        targets.append([0,1,0])
    if match[1]['home_team_goal']<match[1]['away_team_goal']:
        targets.append([0,0,1])
```

```python
teams = data[['home_team_api_id','away_team_api_id']]
```

```python
teams = pd.get_dummies(teams,columns=['home_team_api_id','away_team_api_id'])
```

```python
X_train = teams[:340]
X_test = teams[340:]
y_train = targets[:340]
y_test = targets[340:]
```

```python
from keras.models import Sequential
from keras.layers import Input, Activation, Dense
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(16, input_shape=(40,)))
model.add(Activation("relu"))
model.add(Dense(3))
model.add(Activation("softmax"))

model.compile(loss = "categorical_crossentropy", optimizer = Adam(lr = 1e-3), metrics = ["accuracy"])
model.summary()
```

```python
history = model.fit(X_train, [y_train], batch_size=50, epochs=100, validation_split = 0.1)
```

```python
away_odds = data['PSCA'][340:]
draw_odds = data['PSCD'][340:]
home_odds = data['PSCH'][340:]
```

```python
preds = model.predict(X_test)
```

```python
preds
```

```python
for pred,home,draw,away in zip(preds,home_odds,draw_odds,away_odds):
    print(pred[0]*home)
    print(pred[1]*draw)
    print(pred[2]*away)
    print("------")
```

```python
import numpy as np
y_train= np.array(y_train)
train_results = np.argmax(y_train,axis=1)
y_test = np.array(y_test)
test_results = np.argmax(y_test,axis=1)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,train_results)
```

```python
lr_preds = lr.predict_proba(X_test)
```

```python
lr_preds[2]
```

```python
np.argmax(preds,axis=1)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(np.argmax(preds,axis=1),test_results)
```

```python
for pred,home,draw,away in zip(lr_preds,home_odds,draw_odds,away_odds):
    print(pred[0]*home)
    print(pred[1]*draw)
    print(pred[2]*away)
    print("------")
```

Our model seems to disagree a lot with the odds. Let's see what happens if we bet these odds. Since we'd bet on nearly every match otherwise, let's give our model a bit of a gap and let it bet if it thinks the EV is over 25%

```python
money = 0
bets = 0
for pred,home,draw,away,result in zip(lr_preds,home_odds,draw_odds,away_odds,y_test):
    if (pred[0]*home) > 1.25:
        bets += 1
        money -= 1
        if result[0] == 1:
            money += home
    if (pred[1]*draw) > 1.25:
        bets += 1
        money -= 1
        if result[1] == 1:
            money += draw
    if (pred[2]*away) > 1.25:
        bets += 1
        money -= 1
        if result[2] == 1:
            money += away
```

We made 42 bets of 1 euro and lost a bit over 7 euros with them. I guess it's a start

### Step-by-step imputation

It seems difficult to build a good probability estimate for the result based on just teams. However, it is possible to predict the possession reasonably well. Let's predict the possession and use the predicted value as a feature.

#### Predict possession

```python
possession_df = data[['home_team_api_id','away_team_api_id','possession']]
```

```python
possession_df = pd.get_dummies(possession_df,columns=['home_team_api_id','away_team_api_id'])
```

```python
train = possession_df[:340]
test = possession_df[340:]
```

```python
X_train = train.drop('possession',axis=1)
y_train = train['possession']
X_test = test.drop('possession',axis=1)
y_test = test['possession']
```

Since a neural network seems like a bit of an overkill here, I'll train a simple linear regression model on the data

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
preds = lr.predict(X_test)
```

```python
preds
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(preds,y_test)
mse
```
